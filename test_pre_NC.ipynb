```python
hgt_north_south_slope, hgt_north_south_pvalue = ca.multi_var_regression(hgtERA5_ver_JJA.sel(level=[200.0, 500.0, 850.0]), [preGPCP_nIndia_JJA_nor, preGPCP_sIndia_JJA_nor, preGPCP_NC_JJA_nor], concat_dim="area")
u_north_south_slope, u_north_south_pvalue = ca.multi_var_regression(uERA5_ver_JJA.sel(level=[200.0, 500.0, 850.0]), [preGPCP_nIndia_JJA_nor, preGPCP_sIndia_JJA_nor, preGPCP_NC_JJA_nor], concat_dim="area")
v_north_south_slope, v_north_south_pvalue = ca.multi_var_regression(vERA5_ver_JJA.sel(level=[200.0, 500.0, 850.0]), [preGPCP_nIndia_JJA_nor, preGPCP_sIndia_JJA_nor, preGPCP_NC_JJA_nor], concat_dim="area")
```

```python
hgt_north_south_slope = hgt_north_south_slope.transpose("area", ...)
u_north_south_slope = u_north_south_slope.transpose("area", ...)
v_north_south_slope = v_north_south_slope.transpose("area", ...)

hgt_north_south_pvalue = hgt_north_south_pvalue.transpose("area", ...)
u_north_south_pvalue = u_north_south_pvalue.transpose("area", ...)
v_north_south_pvalue = v_north_south_pvalue.transpose("area", ...)
```
